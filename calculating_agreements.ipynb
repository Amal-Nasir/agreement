```python
allSame = []
threeFourths = []
oneHalf = []
twoRatingsorLess = [] # the zero counts..
noRatings = []
allDifferent = []
aggloSAME = []
agglo75 = []
list2 = []
```

```python
#df = pd.read_csv('/Users/amalalmansour/Desktop/New_Images/Malignancy/machine_ratings/machine_ratings.csv')
#rad_df = pd.read_csv('/Users/amalalmansour/Desktop/New_Images/Malignancy/machine_ratings/Radiologists_ratings.csv')
#f1 = pd.merge(df['noduleID'],rad_df)
df = pd.read_csv('/Users/amalalmansour/Desktop/New_Images/random_agreement/random_numbers30.csv')
display(df)
```

```python
length = 1332 #length of the input file..
counter = 0
for x in range(length):
    
    sp1 = df["m1"][x]
    sp2 = df["m2"][x]
    sp3 = df["m3"][x]
    sp4 = df["m4"][x]

    list2.append(sp1)
    list2.append(sp2)
    list2.append(sp3)
    list2.append(sp4)
    
    zeroCount = list2.count(0)
    #print(zeroCount)   
    
    # All agree... 
    if (sp1 == sp2 == sp3 == sp4):
        counter += 1
        allSame.append(x)
        aggloSAME.append(sp1) # one value from one of the radiologists

    # 75% agree...        
    elif ((sp1 == sp2 == sp3) or (sp1 == sp2 == sp4) or (sp1 == sp3 ==sp4) or (sp2 == sp3 == sp4)):
        threeFourths.append(x)
        #agglo75.append(round((sp1+sp2+sp3+sp4)/4)) # if 1 not spiculated if 0 spiculated
        agglo75.append(statistics.mode([sp1,sp2,sp3,sp4])) # if 1 not spiculated if 0 spiculated

    # 50% agree...
    elif (sp1 == sp2 or sp1 ==sp3 or sp1 == sp4 or sp2 == sp3 or sp2 == sp4 or sp3 == sp4):
        oneHalf.append(x)

    # none agree
    else:
        allDifferent.append(x)
list2.clear()   
```

```python
file_path1 = "/Users/amalalmansour/Desktop/AllAgree.csv"
print(counter)
allsameVal = []
allsameVal.append(df.columns)
for x in allSame:
    #print(allSame.index(x))
    aggloVal = aggloSAME[allSame.index(x)]
    
    df["Agglomeration"][x] = aggloVal
    if(aggloVal == 1):
        df["Binary Rating"][x] = 0 # not spiculated
    else:
        df["Binary Rating"][x] = 1 # spiculated
    arrRow = df.iloc[x]
    arrRow = arrRow.to_numpy()
    allsameVal.append(arrRow)
#np.savetxt(file_path1, allsameVal, fmt='%s', delimiter=',')
print(len(allsameVal))
#print(allsameVal)
```

```python
file_path2 = "/Users/amalalmansour/Desktop/75agree.csv"
three4ths = []
three4ths.append(df.columns)
for x in threeFourths:
    aggloVal = agglo75[threeFourths.index(x)]
    df["Agglomeration"][x] = aggloVal
    if(aggloVal == 1):
        df["Binary Rating"][x] = 0 # not spiculated
    else:
        df["Binary Rating"][x] = 1 # spiculated
    arrRow = df.iloc[x]
    arrRow = arrRow.to_numpy()
    three4ths.append(arrRow)
#np.savetxt(file_path2, three4ths, fmt='%s', delimiter=',')
print(len(three4ths))
```
