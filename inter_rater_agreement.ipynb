```python
import pandas as pd
import numpy as np
from statsmodels.stats import inter_rater as irr
import krippendorff as kd
```

```python
arr = [[1, 0, 5, 4, 2],
       [1, 3, 0, 4, 1],
       [1, 3, 5, 2, 0],
       [0, 3, 4, 4, 5]]

# Remove 0 values from the 2D list
arr_without_zeros = [[value for value in row if value != 0] for row in arr]
arr_without_zeros
```

```python
data = {
    'rater1': [1, 5, 5, 4, 2,3,4],
    'rater2': [1, 4, 5, 4, 2,3,4],
    'rater3': [1, 1, 5, 4, 2,3,4],
    'rater4': [1, 1, 5, 4, 2,3,4]
}

df = pd.DataFrame(data)

# Create new columns for each rating (1-5) and initialize them with zeros
for rating in range(1, 6):
    df[f'column{rating}'] = 0

# Iterate over each row and count the ratings for that row
for i in range(len(df)):
    for rating in range(1, 6):
        count = (df.iloc[i, :-5] == rating).sum()
        df.at[i, f'column{rating}'] = count
        

print(df)
```

```python
data = {
    'rater1': [1, 5, 3, 4, 2],
    'rater2': [1, 5, 3, 4, 2],
    'rater3': [1, 3, 3, 4, 2],
    'rater4': [1, 3, 3, 4, 2]
}

df = pd.DataFrame(data)

# Create new columns for each rating (1-5) and initialize them with zeros
for rating in range(1, 6):
    df[f'column{rating}'] = 0

# Create a 'total' column to store the total count for each row
df['total'] = 0

# Iterate over each row and count the ratings for that row
for i in range(len(df)):
    for rating in range(1, 6):
        count = (df.iloc[i, :-6] == rating).sum()
        df.at[i, f'column{rating}'] = count
    total_count = df.iloc[i, -6:-1].sum()  # Sum the counts for columns 'column1' to 'column5'
    df.at[i, 'total'] = total_count

print(df)
```
